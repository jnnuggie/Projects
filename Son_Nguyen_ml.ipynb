```
steam_df = spark.read.csv("/FileStore/tables/steam_200k.csv", header = False, inferSchema = True)
steam_df.show()
```

```
headings = ["user_id", "game", "action", "hours_played"]

steam_df = steam_df.toDF(*headings)
```

```
steam_df.show()
```

```
from pyspark.sql.functions import countDistinct

highest_games_owned = steam_df.filter(steam_df.action == "purchase") \
                            .groupBy("user_id") \
                            .agg(countDistinct("game").alias("number_of_games")) \
                            .orderBy("number_of_games", ascending = False) \
                            .limit(5)

display(highest_games_owned)
```

```
from pyspark.sql.functions import avg

highest_avg_hours = steam_df.filter(steam_df.action == "play") \
                            .groupBy("game") \
                            .agg(avg("hours_played").alias("average_hours_played")) \
                            .orderBy("average_hours_played", ascending = False) \
                            .limit(10)

display(highest_avg_hours)
```

```
top_avg_hours_result = steam_df.filter(steam_df.game.like("Eastside Hockey M%")) 

top_avg_hours_result.show()
```

```
from pyspark.sql.functions import col

purchased_games = steam_df.filter(col("action") == "purchase").groupBy("game").count()

played_games = steam_df.filter(col("action") == "play").groupBy("game").count()
```

```
purchased_games.show()
```

```
played_games.show()
```

```
purchased_played = purchased_games.join(played_games, ["game"], "left")
```

```
purchased_played.show()
```

```
headings = ["game", "purchased", "played"]

purchased_played = purchased_played.toDF(*headings)
```

```
purchased_played.show()
```

```
from pyspark.sql.functions import coalesce, round, lit

purchased_played = purchased_played.withColumn(
    "percentage_played",
    round((coalesce(col("played"), lit(0)) / col("purchased")) * 100, 2)
)

purchased_played = purchased_played.orderBy(col("percentage_played"), ascending = False)
                        
purchased_played.show()
```

```
from pyspark.sql.functions import monotonically_increasing_id

if "game_id" not in steam_df.columns:
    id_insert = steam_df.select("game").distinct().withColumn("game_id", monotonically_increasing_id())
    steam_df = steam_df.join(id_insert, "game")

steam_df.show()
```

```
steam_df = steam_df.select("user_id", "game_id", "game", "action", "hours_played")

steam_df.show()
```

```
from pyspark.sql.functions import when

steam_df = steam_df.withColumn("rating", when(col("action") == "play", col("hours_played")).otherwise(1))
```

```
(training, test) = steam_df.randomSplit([0.7, 0.3], seed = 100)

training = training.unionAll(steam_df)
```

```
from pyspark.ml.recommendation import ALS

als = ALS (
    maxIter = 5,
    regParam = 0.01,
    userCol = "user_id",
    itemCol = "game_id",
    ratingCol = "rating",
    coldStartStrategy = "drop"
)

model = als.fit(training)
```

```
model = als.fit(training)
```

```
predictions = model.transform(test)
```

```
from pyspark.ml.evaluation import RegressionEvaluator

evaluator = RegressionEvaluator(metricName = "rmse", labelCol = "rating", predictionCol = "prediction")

rmse = evaluator.evaluate(predictions)

print('Root Mean Square Error is %g' %rmse)
```

```
userRecs = model.recommendForAllUsers(10)

userRecs.show(truncate=False)
```

```
user_76767 = steam_df.filter((col("user_id") == 76767) & (col("action") == "play"))

user_76767 = user_76767.select("game", "hours_played")

user_76767.show(truncate=False)
```

```
from pyspark.sql.functions import explode

userRecs.where(userRecs.user_id == 76767).select("recommendations") \
    .withColumn("recommendations", explode("recommendations")) \
    .select("recommendations.game_id", "recommendations.rating") \
    .join(steam_df, ["game_id"]) \
    .show(truncate=False)
```

```
from pyspark.ml.tuning import ParamGridBuilder

parameters = ParamGridBuilder() \
    .addGrid(als.rank, [5, 7, 9]) \
    .addGrid(als.maxIter, [5]) \
    .addGrid(als.regParam, [0.005, 0.01, 0.05, 0.1, 0.5]) \
    .build()
```

```
from pyspark.ml.tuning import CrossValidator

evaluator = RegressionEvaluator(metricName = "rmse", labelCol = "rating", predictionCol = "prediction")

crossval = CrossValidator(
    estimator = als,
    estimatorParamMaps = parameters,
    evaluator = evaluator,
    numFolds = 3
)
```

```
cv_model = crossval.fit(training)
```

```
bestModel = cv_model.bestModel
print("Parameters for the best model: ")
print("Rank Parameter: %g" %bestModel.rank)
print("MaxIter Parameter: %s" %bestModel._java_obj.parent().getMaxIter())
print("RegParam Parameter %g" %bestModel._java_obj.parent().getRegParam())
```

```
userRecs = bestModel.recommendForAllUsers(10)
```
